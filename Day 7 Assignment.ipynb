# 3. Perform Normal, Exponential distributions

```python
import numpy as np
```

```python
mu,sigma=0.9,0.7
s=np.random.normal(mu,sigma,10000)
count,bins,ignored=plt.hist(s,1000)
plt.plot(bins,1/sigma*np.sqrt(2*np.pi)*np.exp(-(bins-mu)**2/(2*sigma**2)))
plt.show()
```

Exponential:

```python
t=np.random.exponential(1,50)
count,bins,ignored=plt.hist(t,8)
plt.show()
```
